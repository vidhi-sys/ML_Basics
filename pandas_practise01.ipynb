**Series Constructor in pandas**:

---


A Pandas Series is a one-dimensional, labeled array-like data structure that can hold data of any type

```python
data=[29,198,87]
series=pd.Series(data,index=[1,2,3])
print(series)
```

```python
print(series.loc[2])
print(series.iloc[0])
# iloc returns by index integers
#loc for location by label
```

```python
print(series[series<100])
```

```python
calories={"day1":420,"day2":480,"day3":1090}
series=pd.Series(calories)
print(series)
```

```python
series.loc['day1']+=80
print(series)
```

```python
print(series[series>=500])
```
